```python
print(df['Gender'].unique())
print(df['City'].unique())
print(df['Category'].unique())
```

```python
df_clean=df.copy()
```

```python
df_clean["Age"] = df_clean["Age"].fillna(df_clean["Age"].median())
```

```python
df_clean["City"] = df_clean["City"].fillna(df_clean["City"].mode()[0])
```

```python
df_clean.isnull().sum()
```

```python
df_clean["Order_Date"] = pd.to_datetime(df_clean["Order_Date"])
```

```python
df_clean.info()
```

```python
df_clean["Order_Month"] = df_clean["Order_Date"].dt.month_name()
```

```python
df_clean["Order_Quarter"] = df_clean["Order_Date"].dt.quarter
```

```python
def age_group(age):
    if age < 25:
        return "Young"
    elif age < 45:
        return "Adult"
    else:
        return "Senior"

df_clean["Age_Group"] = df_clean["Age"].apply(age_group)
```

```python
def revenue_category(sales):
    if sales < 50000:
        return "Low"
    elif sales < 200000:
        return "Medium"
    else:
        return "High"

df_clean["Revenue_Category"] = df_clean["Total_Sales"].apply(revenue_category)
```
